Import Libs

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Creating a Dictionary

```python
data = {
    'date': ['2023/10/18','2023/01/15','2023/07/5','2023/05/15','2023/04/14','2023/06/5','2023/06/26','2023/09/24','2023/01/23','2023/10/20'],
    'MatIce Rating': list(np.random.random(size=10)),
    'KadaIce Rating': list(np.random.random(size=10)),
    'Overall Rating': list(np.random.random(size=10))
}
```

Converting dictionary to a DataFrame

```python
data = pd.DataFrame(data)
```

Create a new column Date

Convert Date from object to datetime and strip off hours

```python
data["Date"] = pd.to_datetime(data["date"])

data["Date"] = data["Date"].dt.date
```

Drop column previous object column date

```python
data.drop('date', axis=1, inplace=True)
```

View the Dataset

```python
data.head(10)
```

Set the column Date as Index

```python
data.set_index('Date', inplace=True)

data
```

View available plot sytle in matplotlib

```python
print(plt.style.available)
plt.style.use('seaborn-v0_8')
```

Plot Visuals
- Line
- Box
- Bar
- Barh
- Pie
- Scatter

```python
data.plot.line(subplots=True, figsize=(13,10), title="Ice Rating",ylabel="Rating")
plt.show()
```

```python
data.plot.box(title='Box Visual')
plt.show()
```

```python
data.plot.bar(figsize=(20,10), title='Bar Visual')
plt.show()
```

```python
data.plot.barh(figsize=(10,10), title='Bar Visual')
plt.show()
```

```python
data.plot.pie(y="MatIce Rating", figsize=(13,10), title='Pie Visual')
plt.show()
```

```python
data.plot.scatter(y='MatIce Rating', x='Overall Rating', title='Scatter Visual')
plt.show()
```
